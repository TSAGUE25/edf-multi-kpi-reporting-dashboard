## 1. Chargement & Vue Parc

```python
df = load_or_generate('../data_sample/edf_kpi_simulated.csv')
print(f'Mois : {df.date.nunique()} | Sites : {df.site.nunique()}')
print(f"Période : {df.date.min()} → {df.date.max()}")
site_kpis, global_kpis = compute_kpis(df)
print('\nKPIs globaux :')
for k, v in global_kpis.items():
    print(f'  {k:<25} : {v:.2f}')
```

## 2. KPIs par Site

```python
print(site_kpis.to_string(index=False))
```

## 3. Mix Énergétique

```python
mix = energy_mix(df)
print('Production par type (GWh) :')
print(mix.round(0))
```

## 4. Alertes Opérationnelles

```python
alerts = alert_sites(df, dispo_threshold=90, incident_threshold=5)
print('Sites en alerte (dernière année) :')
print(alerts[alerts[['alerte_dispo', 'alerte_incidents']].any(axis=1)])
```

## 5. Tendances Mensuelles

```python
monthly = monthly_trend(df)
print('Tendance production (derniers 12 mois) :')
print(monthly.tail(12)[['year','month','production_gwh','revenue_meur','nb_incidents']].to_string(index=False))
```
